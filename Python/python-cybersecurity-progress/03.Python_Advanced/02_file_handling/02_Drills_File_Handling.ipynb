## Testing

Run the cell below. **Don't worry about the `class Test...` syntax** — `unittest` is covered later.


```python
import unittest, os, json, csv

class TestFileHandlingDrills(unittest.TestCase):

    # ---- Part 1: Reading ----

    def test_01_blocked_ips(self):
        self.assertIsInstance(blocked_ips, list,
            "blocked_ips must be a list")
        self.assertGreater(blocked_ip_count, 0,
            "blocked_ip_count must be > 0")
        for ip in blocked_ips:
            self.assertFalse(ip.startswith("#"),
                "No element in blocked_ips should start with # (comment line)")
        self.assertEqual(blocked_ip_count, len(blocked_ips),
            "blocked_ip_count must equal len(blocked_ips)")

    def test_02_status_counts(self):
        self.assertIsInstance(status_counts, dict,
            "status_counts must be a dict")
        self.assertIn("200", status_counts,
            "'200' must be a key — there are successful requests in the log")
        self.assertIn("401", status_counts,
            "'401' must be a key — there are failed login attempts in the log")
        self.assertGreaterEqual(status_counts["401"], 5,
            "There are at least 5 x 401 responses in access.log")

    def test_03_alerts_by_severity(self):
        self.assertIsInstance(alerts_by_severity, dict,
            "alerts_by_severity must be a dict")
        self.assertIn("CRITICAL", alerts_by_severity,
            "'CRITICAL' must be a key")
        self.assertGreaterEqual(len(alerts_by_severity["CRITICAL"]), 2,
            "At least 2 CRITICAL alerts in alerts.txt")
        self.assertGreaterEqual(total_alerts, 8,
            "There are 8 alert lines in alerts.txt")

    # ---- Part 2: Writing ----

    def test_04_repeated_ips(self):
        self.assertTrue(os.path.exists("data/repeated_ips.txt"),
            "data/repeated_ips.txt must be created")
        self.assertGreater(repeated_ip_count, 0,
            "At least one IP appears more than once in access.log")

    def test_05_summary_report(self):
        self.assertTrue(report_exists,
            "data/blocklist_report.txt must exist after writing")
        with open("data/blocklist_report.txt", "r") as f:
            content = f.read()
        self.assertIn("Blocked IP Addresses", content,
            "Report must contain the title")
        self.assertIn("Total: 3 items", content,
            "Report footer must say 'Total: 3 items'")

    def test_06_log_rotation(self):
        self.assertTrue(archive_exists,
            "data/archive/rotated.log must exist after rotation")
        self.assertTrue(current_empty,
            "data/current.log must be 0 bytes after rotation")

    # ---- Part 3: CSV / JSON ----

    def test_07_csv_stats(self):
        self.assertIsInstance(avg_cvss, float,
            "avg_cvss must be a float")
        self.assertGreater(avg_cvss, 0,
            "avg_cvss must be positive")
        self.assertGreater(unpatched_count, 0,
            "unpatched_count must be > 0")
        self.assertIsInstance(highest_cvss_cve, str,
            "highest_cvss_cve must be a string (CVE ID)")
        self.assertTrue(highest_cvss_cve.startswith("CVE-"),
            "highest_cvss_cve must start with 'CVE-'")

    def test_08_unpatched_csv(self):
        self.assertTrue(os.path.exists("data/unpatched_vulns.csv"),
            "data/unpatched_vulns.csv must be created")
        self.assertGreater(unpatched_csv_rows, 0,
            "unpatched_csv_rows must be > 0")
        with open("data/unpatched_vulns.csv", "r", newline="") as f:
            reader = csv.DictReader(f)
            rows = list(reader)
        self.assertEqual(len(rows), unpatched_csv_rows,
            "Row count must match the file content")
        for row in rows:
            self.assertEqual(row["is_patched"], "false",
                "All rows in unpatched_vulns.csv must have is_patched=false")

    def test_09_config_update(self):
        self.assertTrue(config_updated,
            "config_updated must be True")
        self.assertTrue(port_8080_present,
            "Port 8080 must be in firewall.allowed_ports after update")
        with open("data/config.json", "r") as f:
            loaded = json.load(f)
        self.assertIn("last_updated", loaded,
            "'last_updated' key must be present in the updated config")

    # ---- Part 4: Paths ----

    def test_10_pathlib_properties(self):
        self.assertEqual(log_stem, "access",
            "stem of access.log should be 'access'")
        self.assertEqual(log_suffix, ".log",
            "suffix of access.log should be '.log'")
        self.assertGreater(log_size_bytes, 0,
            "log_size_bytes must be > 0")
        self.assertEqual(log_parent_name, "data",
            "parent directory of data/access.log is 'data'")

    def test_11_file_inventory(self):
        self.assertIsInstance(file_inventory, list,
            "file_inventory must be a list")
        self.assertGreater(file_inventory_count, 0,
            "At least one .log or .txt file must be in data/")
        for item in file_inventory:
            self.assertIn("name", item, "Each dict must have 'name'")
            self.assertIn("size_bytes", item, "Each dict must have 'size_bytes'")
            self.assertIn("line_count", item, "Each dict must have 'line_count'")

    def test_12_csv_row_counts(self):
        self.assertIsInstance(csv_row_counts, dict,
            "csv_row_counts must be a dict")
        self.assertIn("vulnerabilities.csv", csv_row_counts,
            "vulnerabilities.csv must be in csv_row_counts")
        self.assertEqual(csv_row_counts["vulnerabilities.csv"], 7,
            "vulnerabilities.csv has 7 data rows (excluding header)")

    # ---- Part 5: Putting it together ----

    def test_13_threat_ips(self):
        self.assertIsInstance(threat_log_lines, list,
            "threat_log_lines must be a list")
        self.assertGreater(threat_hit_count, 0,
            "At least one log line should match a blocked IP")
        self.assertEqual(threat_hit_count, len(threat_log_lines),
            "threat_hit_count must equal len(threat_log_lines)")

    def test_14_incident_report(self):
        self.assertTrue(os.path.exists("data/incident_report.json"),
            "data/incident_report.json must exist")
        self.assertGreater(report_total_requests, 0,
            "report_total_requests must be > 0")
        with open("data/incident_report.json", "r") as f:
            report = json.load(f)
        required_keys = ["total_requests", "unique_ips", "failed_auth_401",
                         "forbidden_403", "top_ip", "top_ip_requests"]
        for key in required_keys:
            self.assertIn(key, report, f"Report must contain key: '{key}'")

    def test_15_pipeline(self):
        self.assertGreater(pipeline_flagged_count, 0,
            "Pipeline must flag at least one event")
        self.assertEqual(
            pipeline_flagged_count,
            pipeline_blocklist_hits + pipeline_auth_failures,
            "Total flagged = blocklist hits + auth failures (no double-counting required)"
        )
        self.assertTrue(os.path.exists("data/flagged_events.csv"),
            "data/flagged_events.csv must be created")
        self.assertTrue(os.path.exists("data/pipeline_summary.json"),
            "data/pipeline_summary.json must be created")


unittest.main(argv=[""], verbosity=2, exit=False)
```
