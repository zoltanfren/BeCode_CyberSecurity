## Testing

Run the cell below to check all your answers.
**Don't worry about the `class Test...` syntax** — this is `unittest`, covered later in the curriculum.


````python
import unittest

class TestExceptionHandling(unittest.TestCase):

    # ---- Part 1: try/except ----

    def test_01_safe_int_valid(self):
        self.assertEqual(result_valid, 42,
            "safe_int('42') should return the integer 42")

    def test_02_safe_int_invalid(self):
        self.assertIsNone(result_str,
            "safe_int('hello') should return None — ValueError must be caught")
        self.assertIsNone(result_none,
            "safe_int(None) should return None — TypeError must be caught")

    def test_03_safe_lookup(self):
        self.assertEqual(lookup_hit, "SSH",
            "safe_lookup(port_services, 22) should return 'SSH'")
        self.assertIsNone(lookup_miss,
            "safe_lookup for a missing key should return None")
        self.assertIsNone(lookup_type_error,
            "safe_lookup with an unhashable key should return None")

    def test_04_parse_port(self):
        self.assertEqual(port_valid, 443,
            "parse_port('443') should return the integer 443")
        self.assertIsNone(port_zero,
            "parse_port('0') should return None — port 0 is out of range")
        self.assertIsNone(port_bad_str,
            "parse_port('https') should return None — cannot convert to int")

    # ---- Part 2: else / finally ----

    def test_05_read_severity(self):
        self.assertEqual(sev_critical, "CRITICAL",
            "Score 9.8 should map to CRITICAL")
        self.assertEqual(sev_medium, "MEDIUM",
            "Score 5.5 should map to MEDIUM")
        self.assertEqual(sev_unknown, "UNKNOWN",
            "Non-numeric input should return 'UNKNOWN'")

    def test_06_attempt_connection(self):
        self.assertIsInstance(conn_local, bool,
            "attempt_connection must return a bool")
        self.assertIsInstance(conn_bad, bool,
            "attempt_connection must return a bool even for bad hosts")
        self.assertFalse(conn_bad,
            "conn_bad should be False — invalid host cannot connect")

    def test_07_safe_divide(self):
        self.assertEqual(div_ok, 5.0,
            "safe_divide(10, 2) should return 5.0")
        self.assertIsNone(div_zero,
            "Division by zero should return None")
        self.assertIsNone(div_type,
            "Division by string should return None")

    # ---- Part 3: raise ----

    def test_08_validate_ip(self):
        self.assertTrue(ip_valid,
            "'192.168.1.1' is a valid IP address")
        self.assertTrue(ip_bad_format,
            "'192.168.1' has only 3 octets — must raise ValueError")
        self.assertTrue(ip_bad_type,
            "An integer is not a valid IP — must raise TypeError")

    def test_09_set_alert_severity(self):
        self.assertEqual(sev_ok, "CRITICAL",
            "'CRITICAL' is a valid severity — should be returned as-is")
        self.assertTrue(sev_caught,
            "'URGENT' is not a valid severity — must raise ValueError")

    def test_10_process_scan_result(self):
        self.assertIsInstance(scan_ok, str,
            "process_scan_result should return a string on success")
        self.assertIn("10.0.0.5", scan_ok,
            "The result string must include the target IP")
        self.assertTrue(scan_type_err,
            "Passing a list should raise TypeError")
        self.assertTrue(scan_key_err,
            "Missing 'open_ports' key should raise KeyError")

    # ---- Part 4: Custom Exceptions ----

    def test_11_invalid_port_error(self):
        self.assertEqual(conn_ok, "Connecting on port 443",
            "open_connection(443) should return the connection string")
        self.assertEqual(caught_port, 0,
            "caught_port should be 0 — the invalid port that was passed")

    def test_12_scan_error_hierarchy(self):
        self.assertIsInstance(scan_ok, str,
            "simulate_scan with no error should return a string")
        self.assertTrue(caught_unreachable,
            "simulate_scan('10.0.0.2', 'unreachable') should raise TargetUnreachableError")
        self.assertTrue(caught_timeout,
            "simulate_scan('10.0.0.3', 'timeout') should raise ScanTimeoutError")

    def test_13_policy_violation_error(self):
        self.assertTrue(policy_ok,
            "'Str0ngPassw0rd!' meets all policy requirements")
        self.assertTrue(policy_caught,
            "'weak' should raise PolicyViolationError")
        self.assertEqual(caught_rule, "MIN_LENGTH",
            "'weak' fails MIN_LENGTH first — rule_name should be 'MIN_LENGTH'")

    # ---- Part 5: Putting It Together ----

    def test_14_ingest_alert(self):
        self.assertIsInstance(alert_ok, dict,
            "A valid line should return a dict")
        self.assertEqual(alert_ok["alert_id"], "ALT-0042",
            "alert_id should be 'ALT-0042'")
        self.assertEqual(alert_ok["cvss_score"], 9.5,
            "cvss_score should be the float 9.5")
        self.assertIsNone(alert_bad_fields,
            "A line with only 2 fields should return None after catching ValueError")
        self.assertIsNone(alert_bad_score,
            "A score of 15.0 is out of range — should return None")

    def test_15_run_vulnerability_check(self):
        self.assertIsInstance(check_ok, dict,
            "A valid check should return a dict")
        self.assertEqual(check_ok["severity"], "CRITICAL",
            "CVSS 10.0 should be CRITICAL")
        self.assertTrue(check_bad_cve_caught,
            "'MS-001' is not a CVE ID — must raise ValueError")

    def test_16_pipeline(self):
        self.assertIsInstance(pipeline_results, list,
            "process_alerts must return a list")
        self.assertEqual(pipeline_success_count, 2,
            "2 of the 4 lines are valid — broken line and score 99.9 must be rejected")


unittest.main(argv=[""], verbosity=2, exit=False)
````
