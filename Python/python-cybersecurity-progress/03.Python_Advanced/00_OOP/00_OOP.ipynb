## 1. What is OOP? — The Blueprint Analogy

Imagine you are a security engineer and you need to track **vulnerabilities**.
Every vulnerability has the same kind of information: a CVE ID, a severity score, an affected system,
and whether it has been patched.

You could create 4 separate variables for each vulnerability... but you have 200 vulnerabilities.
That is 800 variables. Chaos.

**Object-Oriented Programming (OOP)** solves this by letting you define a **blueprint** — called a **class** —
that describes what a vulnerability looks like. Then you create as many **objects** (instances) from that
blueprint as you need, each with its own data.

```
CLASS (Blueprint)         OBJECT (Real instance)
------------------        --------------------------------
Vulnerability  ------>    CVE-2023-44487  (HTTP/2 Rapid Reset)
               ------>    CVE-2021-44228  (Log4Shell)
               ------>    CVE-2023-23397  (Outlook RCE)
```

The blueprint defines the **structure**. The objects hold the **real data**.

### Key vocabulary

| Term | Plain English | Example |
|------|-------------|--------|
| **Class** | The blueprint / template | `Vulnerability` |
| **Object / Instance** | One real thing built from the blueprint | `CVE-2021-44228` |
| **Attribute** | A piece of data that belongs to the object | `cvss_score = 10.0` |
| **Method** | A function that belongs to the object | `is_critical()` |

## 2. Your First Class

The syntax to define a class is:

```python
class ClassName:
    """Docstring: describe what this class represents"""
    pass  # placeholder — we'll add content soon
```

Rules:
- Class names use **PascalCase** (each word capitalised, no underscores): `Vulnerability`, `FirewallRule`, `LogEntry`
- Functions use `snake_case` — classes are the exception
- The `class` keyword is followed by the name and a colon
- Everything inside is indented (same rule as `if` and `def`)


````python
# Our first class — just the shell
# We'll fill it in over the next sections

class Vulnerability:
    """Represents a security vulnerability tracked by the SOC team."""
    pass  # empty for now


# Creating an object (instance) from the class:
vuln_1 = Vulnerability()  # Vulnerability() calls the class like a function
vuln_2 = Vulnerability()  # a completely separate object

print(type(vuln_1))   # confirms it is a Vulnerability object
print(vuln_1 is vuln_2)  # False — they are two different objects in memory
````

## 3. The Constructor — `__init__`

Right now our class is an empty shell. We need a way to give each object its own data
when it is created. That is the job of the **constructor**: a special method called `__init__`.

`__init__` runs **automatically** every time you create a new object.

```python
class ClassName:
    def __init__(self, param1, param2):
        self.param1 = param1   # store as attribute on this object
        self.param2 = param2
```

### What is `self`?

Think of `self` as **the object referring to itself**. When you write `self.cvss_score`, you are saying:
"attach the attribute `cvss_score` to *this specific object*".

- `self` is **always the first parameter** in any method — but you never pass it yourself
- Python passes it automatically when you call the method

```
Vulnerability("CVE-2021-44228", 10.0)
      |                |         |
    class             cve_id   cvss_score
                      (self is handled automatically)
```


````python
# Topic: The __init__ constructor — giving each object its own data

class Vulnerability:
    """Represents a security vulnerability tracked by the SOC team."""

    def __init__(self, cve_id, cvss_score, affected_system, is_patched):
        """Called automatically when a Vulnerability object is created."""
        self.cve_id = cve_id                 # e.g. 'CVE-2021-44228'
        self.cvss_score = cvss_score          # float 0.0-10.0
        self.affected_system = affected_system  # e.g. 'Apache Log4j'
        self.is_patched = is_patched          # True/False


# Creating two vulnerability objects — each holds its own separate data
log4shell = Vulnerability("CVE-2021-44228", 10.0, "Apache Log4j", False)
outlook_rce = Vulnerability("CVE-2023-23397", 9.8, "Microsoft Outlook", True)

# Accessing attributes with dot notation
print(log4shell.cve_id)          # CVE-2021-44228
print(log4shell.cvss_score)      # 10.0
print(outlook_rce.is_patched)    # True
````

## 4. Instance Attributes

Every attribute you set with `self.something = value` is an **instance attribute** — it belongs to
that specific object and is completely independent from other objects.

You can also **modify** attributes after creation using dot notation:

```python
object.attribute = new_value
```

This is useful in security workflows — for example, when you patch a vulnerability, you update
`is_patched` from `False` to `True`.


````python
# Topic: Reading and modifying instance attributes

class Vulnerability:
    def __init__(self, cve_id, cvss_score, affected_system, is_patched):
        self.cve_id = cve_id
        self.cvss_score = cvss_score
        self.affected_system = affected_system
        self.is_patched = is_patched


log4shell = Vulnerability("CVE-2021-44228", 10.0, "Apache Log4j", False)

print("Before patch:", log4shell.is_patched)   # False

# The patch has been applied — update the attribute
log4shell.is_patched = True

print("After patch: ", log4shell.is_patched)   # True

# You can also update numeric attributes
# (e.g. CVSS score revised by NVD after initial publication)
log4shell.cvss_score = 10.0   # confirmed critical
print("CVSS score:  ", log4shell.cvss_score)
````

## 5. Methods — What Can the Object Do?

Attributes describe **what an object IS**. Methods describe **what an object can DO**.

A method is just a function defined inside a class. The only rule: it must always
have `self` as its first parameter — this gives it access to the object's attributes.

```python
class ClassName:
    def __init__(self, ...):
        self.attribute = value

    def my_method(self):      # <-- self gives access to attributes
        # do something with self.attribute
        return result
```

Methods are called using **dot notation** on the object:

```python
result = my_object.my_method()
```


````python
# Topic: Adding methods to a class

class Vulnerability:
    """Represents a security vulnerability tracked by the SOC team."""

    def __init__(self, cve_id, cvss_score, affected_system, is_patched):
        self.cve_id = cve_id
        self.cvss_score = cvss_score
        self.affected_system = affected_system
        self.is_patched = is_patched

    def get_severity(self):
        """Returns severity label based on CVSS score (NIST scale)."""
        if self.cvss_score >= 9.0:
            return "CRITICAL"
        elif self.cvss_score >= 7.0:
            return "HIGH"
        elif self.cvss_score >= 4.0:
            return "MEDIUM"
        else:
            return "LOW"

    def needs_urgent_action(self):
        """Returns True if the vuln is CRITICAL or HIGH and not yet patched."""
        is_severe = self.cvss_score >= 7.0   # HIGH or CRITICAL
        return is_severe and not self.is_patched


# Testing our methods
log4shell = Vulnerability("CVE-2021-44228", 10.0, "Apache Log4j", False)
low_risk = Vulnerability("CVE-2023-00001", 3.2, "Internal tool", False)

print(log4shell.get_severity())          # CRITICAL
print(log4shell.needs_urgent_action())   # True  (critical + not patched)
print(low_risk.needs_urgent_action())    # False (low severity)
````

## 6. The `__str__` Method — Human-Readable Output

Right now, if you print a `Vulnerability` object, Python shows something like:
`<__main__.Vulnerability object at 0x7f3a2b...>` — completely unreadable.

The `__str__` method lets you define what Python should print when you do `print(my_object)`.
It must always return a **string**.

```python
def __str__(self):
    return f"some string with {self.attributes}"
```

This is used for example for log formatters, alert summaries,
vulnerability reports all rely on clean string representations of objects.


````python
# Topic: __str__ for readable security tool output

class Vulnerability:
    def __init__(self, cve_id, cvss_score, affected_system, is_patched):
        self.cve_id = cve_id
        self.cvss_score = cvss_score
        self.affected_system = affected_system
        self.is_patched = is_patched

    def get_severity(self):
        if self.cvss_score >= 9.0:
            return "CRITICAL"
        elif self.cvss_score >= 7.0:
            return "HIGH"
        elif self.cvss_score >= 4.0:
            return "MEDIUM"
        else:
            return "LOW"

    def __str__(self):
        """Returns a one-line summary suitable for a SOC report."""
        patch_status = "PATCHED" if self.is_patched else "UNPATCHED"
        return (
            f"[{self.get_severity()}] {self.cve_id} | "
            f"CVSS: {self.cvss_score} | "
            f"System: {self.affected_system} | "
            f"Status: {patch_status}"
        )


log4shell = Vulnerability("CVE-2021-44228", 10.0, "Apache Log4j", False)
print(log4shell)
# [CRITICAL] CVE-2021-44228 | CVSS: 10.0 | System: Apache Log4j | Status: UNPATCHED
````

## 7. Putting It Together — A Security Tool Object

Let's build a more complete object that combines everything: constructor, attributes, multiple methods, and `__str__`.

This is the kind of class you would write in a real SOC tool.


````python
# Topic: Full class combining all ELI5 concepts
# Real-world scenario: a basic alert object for a SIEM-like tool

class SecurityAlert:
    """Represents a security alert raised by detection rules."""

    def __init__(self, alert_id, source_ip, rule_name, severity, is_acknowledged):
        self.alert_id = alert_id              # unique identifier
        self.source_ip = source_ip            # attacker or suspicious IP
        self.rule_name = rule_name            # detection rule that fired
        self.severity = severity              # 'LOW', 'MEDIUM', 'HIGH', 'CRITICAL'
        self.is_acknowledged = is_acknowledged  # analyst has seen it

    def acknowledge(self):
        """Mark this alert as seen by an analyst."""
        self.is_acknowledged = True

    def is_high_priority(self):
        """Returns True if this alert needs immediate analyst attention."""
        urgent_levels = ["HIGH", "CRITICAL"]     # LOW and MEDIUM can wait
        return self.severity in urgent_levels and not self.is_acknowledged

    def __str__(self):
        status = "ACK" if self.is_acknowledged else "NEW"
        return (
            f"[{status}][{self.severity}] Alert #{self.alert_id} | "
            f"Rule: {self.rule_name} | Source: {self.source_ip}"
        )


# Simulate a new brute-force alert arriving
alert = SecurityAlert(
    alert_id="ALT-0042",
    source_ip="185.220.101.47",
    rule_name="SSH_BRUTE_FORCE",
    severity="HIGH",
    is_acknowledged=False
)

print(alert)
print("High priority:", alert.is_high_priority())   # True

alert.acknowledge()
print(alert)
print("High priority:", alert.is_high_priority())   # False (acknowledged)
````
