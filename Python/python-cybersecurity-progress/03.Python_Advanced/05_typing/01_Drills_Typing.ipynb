## Test Suite

Run the cell below after completing all exercises.  
Do not worry about the `class` / `unittest` syntax - it will be covered in a later module.
Focus on the output: each passing test shows `OK`, each failure tells you what went wrong.


```python
import unittest
from typing import Dict, List, Optional, Union, Tuple


class TestTypingDrills(unittest.TestCase):

    # --- Part 1: Variable Annotations ---

    def test_01_variable_annotations_types(self):
        """Exercise 1: all five variables must hold the correct values."""
        self.assertEqual(target_ip, "10.0.0.254",
                         "target_ip should be '10.0.0.254' (str)")
        self.assertEqual(ssh_port, 22,
                         "ssh_port should be 22 (int)")
        self.assertAlmostEqual(scan_timeout, 3.5,
                               msg="scan_timeout should be 3.5 (float)")
        self.assertTrue(vpn_active,
                        "vpn_active should be True (bool)")
        self.assertIsInstance(vpn_active, bool,
                              "vpn_active must be bool, not int")
        self.assertEqual(max_retries, 3,
                         "max_retries should be 3 (int)")

    def test_02_fixed_annotations_values(self):
        """Exercise 2: after fixing annotations, values must remain unchanged."""
        self.assertAlmostEqual(cvss_score, 9.8,
                               msg="cvss_score must be 9.8 (float) - annotation was wrong type")
        self.assertIsInstance(cvss_score, float,
                              "cvss_score must be float, not int")
        self.assertEqual(cve_id, "CVE-2024-1234",
                         "cve_id must be 'CVE-2024-1234' (str) - annotation was wrong type")
        self.assertIsInstance(cve_id, str,
                              "cve_id must be str, not int")
        self.assertTrue(firewall_enabled,
                        "firewall_enabled must be True (bool) - annotation was wrong type")
        self.assertIsInstance(firewall_enabled, bool,
                              "firewall_enabled must be bool, not str")
        self.assertIsInstance(open_ports, list,
                              "open_ports must be a list, not str")
        self.assertEqual(open_ports, [22, 80, 443],
                         "open_ports must be [22, 80, 443] - annotation was wrong type")

    def test_03_declared_variables(self):
        """Exercise 3: four variables must be declared with correct values."""
        self.assertEqual(incident_id, "INC-2024-0099",
                         "incident_id should be 'INC-2024-0099'")
        self.assertIsInstance(incident_id, str,
                              "incident_id must be a str")
        self.assertEqual(affected_hosts, 7,
                         "affected_hosts should be 7")
        self.assertIsInstance(affected_hosts, int,
                              "affected_hosts must be int")
        self.assertFalse(is_resolved,
                         "is_resolved should be False")
        self.assertIsInstance(is_resolved, bool,
                              "is_resolved must be bool, not int")
        self.assertAlmostEqual(response_time_hours, 2.5,
                               msg="response_time_hours should be 2.5")
        self.assertIsInstance(response_time_hours, float,
                              "response_time_hours must be float")

    # --- Part 2: Function Annotations ---

    def test_04_is_port_open(self):
        """Exercise 4: is_port_open must return correct booleans."""
        self.assertTrue(is_port_open("10.0.0.1", 22),
                        "Port 22 < 1024 - should return True")
        self.assertFalse(is_port_open("10.0.0.1", 8080),
                         "Port 8080 >= 1024 - should return False")
        self.assertIsInstance(is_port_open("10.0.0.1", 22), bool,
                              "is_port_open must return bool")

    def test_05_classify_severity(self):
        """Exercise 5: classify_severity must return correct labels."""
        self.assertEqual(classify_severity(9.8), "CRITICAL",
                         "CVSS 9.8 should map to 'CRITICAL' (score >= 9.0)")
        self.assertEqual(classify_severity(7.5), "HIGH",
                         "CVSS 7.5 should map to 'HIGH' (7.0 <= score < 9.0)")
        self.assertEqual(classify_severity(5.0), "MEDIUM",
                         "CVSS 5.0 should map to 'MEDIUM' (4.0 <= score < 7.0)")
        self.assertEqual(classify_severity(2.0), "LOW",
                         "CVSS 2.0 should map to 'LOW' (score < 4.0)")
        self.assertIsInstance(classify_severity(9.8), str,
                              "classify_severity must return a str")

    def test_06_log_event_output(self):
        """Exercise 6: log_event must return None (side-effect function)."""
        result = log_event("10.0.0.5", "port_scan", 22, blocked=True)
        self.assertIsNone(result,
                          "log_event returns None - it only prints, no return value")

    def test_07_fixed_functions(self):
        """Exercise 7: after fixing annotations, functions must work correctly."""
        self.assertTrue(check_threshold(9.8),
                        "check_threshold(9.8) should return True (9.8 >= 7.0 default threshold)")
        self.assertFalse(check_threshold(5.0),
                         "check_threshold(5.0) should return False (5.0 < 7.0)")
        self.assertIsInstance(check_threshold(9.8), bool,
                              "check_threshold must return bool")
        result = summarise_host("10.0.0.1", 3)
        self.assertIsInstance(result, str,
                              "summarise_host must return a str - check return annotation")
        self.assertIn("10.0.0.1", result,
                      "summarise_host result must include the IP address")

    # --- Part 3: Complex Types ---

    def test_08_complex_variable_types(self):
        """Exercise 8: all six variables must have correct types and values."""
        self.assertIsInstance(blocked_ips, list,
                              "blocked_ips must be a list")
        self.assertEqual(blocked_ips, ["203.0.113.5", "198.51.100.42"],
                         "blocked_ips values must not change")
        self.assertIsInstance(open_ports, list,
                              "open_ports must be a list")
        self.assertIsInstance(port_services, dict,
                              "port_services must be a dict")
        self.assertEqual(port_services[22], "SSH",
                         "port_services[22] should be 'SSH'")
        self.assertIsInstance(cve_scores, dict,
                              "cve_scores must be a dict")
        self.assertIsInstance(ssh_endpoint, tuple,
                              "ssh_endpoint must be a tuple, not a list")
        self.assertEqual(ssh_endpoint, ("10.0.0.1", 22),
                         "ssh_endpoint should be ('10.0.0.1', 22) - a (ip, port) tuple")
        self.assertIsInstance(active_ips, set,
                              "active_ips must be a set, not a list")

    def test_09_filter_critical_cves(self):
        """Exercise 9: filter_critical_cves must return only CVEs at or above threshold."""
        test_db = {"CVE-A": 9.8, "CVE-B": 7.5, "CVE-C": 9.1, "CVE-D": 4.0}
        result = filter_critical_cves(test_db)
        self.assertIsInstance(result, list,
                              "filter_critical_cves must return a list")
        self.assertIn("CVE-A", result,
                      "CVE-A (9.8) should be in results for threshold=9.0")
        self.assertIn("CVE-C", result,
                      "CVE-C (9.1) should be in results for threshold=9.0")
        self.assertNotIn("CVE-B", result,
                         "CVE-B (7.5) should NOT be in results for threshold=9.0")
        self.assertNotIn("CVE-D", result,
                         "CVE-D (4.0) should NOT be in results for threshold=9.0")
        # Custom threshold
        high_and_above = filter_critical_cves(test_db, 7.0)
        self.assertIn("CVE-B", high_and_above,
                      "CVE-B (7.5) SHOULD be included when threshold=7.0")

    def test_10_build_port_map(self):
        """Exercise 10: build_port_map must correctly group ports by IP."""
        raw = [("10.0.0.1", 22), ("10.0.0.1", 443), ("10.0.0.2", 80)]
        result = build_port_map(raw)
        self.assertIsInstance(result, dict,
                              "build_port_map must return a dict")
        self.assertIn("10.0.0.1", result,
                      "10.0.0.1 should be a key in the result")
        self.assertEqual(sorted(result["10.0.0.1"]), [22, 443],
                         "10.0.0.1 should have ports [22, 443]")
        self.assertEqual(result["10.0.0.2"], [80],
                         "10.0.0.2 should have ports [80]")

    # --- Part 4: Optional and Union ---

    def test_11_lookup_service(self):
        """Exercise 11: lookup_service must return a str when found, None when not."""
        services = {22: "SSH", 80: "HTTP", 443: "HTTPS"}
        self.assertEqual(lookup_service(22, services), "SSH",
                         "lookup_service(22) should return 'SSH'")
        self.assertIsNone(lookup_service(9999, services),
                          "lookup_service(9999) should return None - port not in map")

    def test_12_safe_parse_port(self):
        """Exercise 12: safe_parse_port must parse valid numbers, return None otherwise."""
        self.assertEqual(safe_parse_port("443"), 443,
                         "'443' is a valid integer string - should return 443")
        self.assertIsInstance(safe_parse_port("443"), int,
                              "safe_parse_port must return int, not str")
        self.assertIsNone(safe_parse_port("HTTPS"),
                          "'HTTPS' cannot be parsed as int - should return None")
        self.assertIsNone(safe_parse_port("9.8"),
                          "'9.8' cannot be parsed as int (it's a float) - should return None")

    def test_13_format_threat_source(self):
        """Exercise 13: format_threat_source must handle both str and int inputs."""
        ip_result = format_threat_source("203.0.113.5")
        self.assertIn("203.0.113.5", ip_result,
                      "IP result must contain the IP address")
        self.assertIn("IP", ip_result,
                      "IP result should contain the word 'IP'")
        port_result = format_threat_source(4444)
        self.assertIn("4444", port_result,
                      "Port result must contain '4444'")
        self.assertIn("Port", port_result,
                      "Port result should contain the word 'Port'")
        self.assertIsInstance(ip_result, str,
                              "format_threat_source must return str in both cases")
        self.assertIsInstance(port_result, str,
                              "format_threat_source must return str in both cases")

    # --- Part 5: Putting It Together ---

    def test_14_count_exposed_ports(self):
        """Exercise 14: count_exposed_ports must sum all ports across all hosts."""
        scan = {
            "10.0.0.1": [22, 80, 443],
            "10.0.0.2": [22, 3389],
            "10.0.0.3": [8080]
        }
        self.assertEqual(count_exposed_ports(scan), 6,
                         "Total ports: 3 + 2 + 1 = 6")
        self.assertIsInstance(count_exposed_ports(scan), int,
                              "count_exposed_ports must return int")
        self.assertEqual(count_exposed_ports({}), 0,
                         "Empty network map should return 0")

    def test_15_generate_incident_report(self):
        """Exercise 15: generate_incident_report must return correct dict structure."""
        report = generate_incident_report(
            host_ip="10.0.0.55",
            open_ports=[22, 3389, 445],
            vuln_db={"CVE-2024-1234": 9.8, "CVE-2023-9876": 7.5},
            analyst_note="RDP exposed - urgent"
        )
        self.assertIsInstance(report, dict,
                              "generate_incident_report must return a dict")
        self.assertEqual(report["host"], "10.0.0.55",
                         "'host' key should contain the IP address")
        self.assertEqual(report["port_count"], 3,
                         "'port_count' should be 3 (three open ports)")
        self.assertIsInstance(report["critical_cves"], list,
                              "'critical_cves' must be a list")
        self.assertIn("CVE-2024-1234", report["critical_cves"],
                      "CVE-2024-1234 (score 9.8) should be in critical_cves")
        self.assertNotIn("CVE-2023-9876", report["critical_cves"],
                         "CVE-2023-9876 (score 7.5) should NOT be in critical_cves (threshold 9.0)")
        self.assertEqual(report["note"], "RDP exposed - urgent",
                         "'note' should contain the analyst note when provided")
        # Test without note
        report_no_note = generate_incident_report("10.0.0.1", [22], {})
        self.assertEqual(report_no_note["note"], "No note",
                         "'note' should be 'No note' when analyst_note is None")


unittest.main(argv=[""], verbosity=2, exit=False)
```
